```python
import pandas as pd
import numpy as np

# get website cookie data
cookies_df = pd.read_json('OpenWPM-proj/data.json').T
cookies_df.head()
```

```python
# get 25th, 75th percentile for numeric fields
cookies_25 = np.percentile(cookies_df["num_cookies"], 25)
cookies_75 = np.percentile(cookies_df["num_cookies"], 75)
not_http_25 = np.percentile(cookies_df["not_http"], 25)
not_http_75 = np.percentile(cookies_df["not_http"], 75)
not_host_25 = np.percentile(cookies_df["not_host"], 25)
not_host_75 = np.percentile(cookies_df["not_host"], 75)

print("cookies: ", cookies_25, cookies_75)
print("not_http: ", not_http_25, not_http_75)
print("not_host: ", not_host_25, not_host_75)
```

```python
cookies_df["num_cookies"] = pd.to_numeric(cookies_df["num_cookies"])
cookies_df["not_http"] = pd.to_numeric(cookies_df["not_http"])
cookies_df["not_host"] = pd.to_numeric(cookies_df["not_host"])

cookies_numeric = cookies_df[["num_cookies", "not_http", "not_host"]]
hist_cookies_df = pd.DataFrame(cookies_numeric)

hist = hist_cookies_df.hist(bins=100, figsize = (16,10))
```
